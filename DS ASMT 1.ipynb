```python
df['age'].fillna(df['age'].mean(),axis = 0,inplace = True)
df.drop('deck',axis = 1,inplace = True)
cols = ['embarked','embark_town']
for col in cols:
    df[col].fillna(df[col].mode()[0],inplace = True)
```

```python
df['age'] = df['age'].astype('int')
df
```

```python
df = pd.get_dummies(df,columns=['who'])
df
```

```python
# Normalize the 'Fare' column using Min-Max scaling
df['Fare_normalized'] = (df['fare'] - df['fare'].min()) / (df['fare'].max() - df['fare'].min())
```
