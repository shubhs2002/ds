Import Libraries (Pandas)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
df = pd.read_csv("covid_vaccine_statewise.csv")
```

a. Describe the dataset

```python
print(df.head())
```

```python
print(df.tail())
```

```python
print("Dataset Description:")
print(df.describe())
```

b. Number of persons state wise vaccinated for first dose in India

```python
first_dose_vaccinated = df.groupby('State')['First Dose Administered'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=first_dose_vaccinated.values, y=first_dose_vaccinated.index, palette='viridis')
plt.title('Number of persons state wise vaccinated for first dose in India')
plt.xlabel('Number of Persons Vaccinated')
plt.ylabel('State/UT')
plt.show()
```

c. Number of persons state wise vaccinated for second dose in India

```python
second_dose_vaccinated = df.groupby('State')['Second Dose Administered'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=second_dose_vaccinated.values, y=second_dose_vaccinated.index, palette='viridis')
plt.title('Number of persons state wise vaccinated for second dose in India')
plt.xlabel('Number of Persons Vaccinated')
plt.ylabel('State/UT')
plt.show()
```

d. Number of Males vaccinated

```python
males_vaccinated = df['Male(Individuals Vaccinated)'].sum()
print("\nNumber of Males vaccinated in India:", males_vaccinated)
```

e. Number of Females vaccinated

```python
females_vaccinated = df['Female(Individuals Vaccinated)'].sum()
print("\nNumber of Females vaccinated in India:", females_vaccinated)
```

Creating a pie chart for gender-wise vaccination distribution

```python
plt.figure(figsize=(8, 6))
labels = ['Male', 'Female']
sizes = [males_vaccinated, females_vaccinated]
colors = ['#ff9999','#66b3ff']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Gender-wise Vaccination Distribution in India')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
